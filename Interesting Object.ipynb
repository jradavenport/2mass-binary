```python
# get the pythons rolling
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

```python
file_2mass = 'IRSA-data/pswdb-31046.lc'
```

```python
data = np.genfromtxt(file_2mass, skip_header=112)
```

```python
# orbital period, found from old IDL analysis
per = 2.034938711 

Jband = data[:,6]
Hband = data[:,14]
Kband = data[:,22]
time = data[:,3]
```

```python
f, axarr = plt.subplots(2, sharex=True, figsize=(8,5))

axarr[0].scatter(np.mod(time, per) / per, Jband,marker='x')
axarr[0].set_ylabel('J band (mag)')
axarr[0].set_ylim((Jband.max(), Jband.min()))
axarr[0].set_xlim((0,1))

axarr[1].scatter(np.mod(time, per) / per, Jband-Kband, marker='+',c='red')
axarr[1].set_ylabel('J-K color (mag)')
axarr[1].set_xlabel('Phase')
axarr[1].set_xlim((0,1))
plt.savefig('binary.png')

```
